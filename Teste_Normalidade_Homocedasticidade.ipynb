#Testes de Normalidade - Em construção

Antes de aplicar determinados métodos estatísticos, é necessário avaliar como os dados estão distribuídos, verificando se seguem uma distribuição normal.

A distribuição normal é um tipo de distribuição em que os dados se concentram em torno de um valor central, enquanto a frequência das observações diminui gradualmente à medida que se afasta desse centro, formando uma curva simétrica em forma de sino. Nela, a média, a mediana e a moda coincidem, indicando equilíbrio dos dados,

![Distribuicao Normal](https://media.datacamp.com/cms/8e0c4782252bd852f6c719e5e9d09c37.png)
Fonte: https://media.datacamp.com/cms/8e0c4782252bd852f6c719e5e9d09c37.png

No entanto, nem todos os conjuntos de dados seguem esse padrão, tornando necessária a verificação da normalidade antes da aplicação de determinados métodos estatísticos. Existem duas formas principais para verificar a condição de normalidade: métodos gráficos e testes estatísticos, que serão apresentados a seguir.

### Métodos gráficos

**Histograma:**
O histograma é um gráfico que permite visualizar a distribuição dos dados. Quando os dados seguem uma distribuição normal, o histograma apresenta formato semelhante a uma curva em sino, com maior concentração de observações no centro e distribuição aproximadamente simétrica em torno da média.

![Histograma](https://media.datacamp.com/cms/ac4b882626f774b45a8335d090ee3d7b.png)
Fonte: https://media.datacamp.com/cms/ac4b882626f774b45a8335d090ee3d7b.png

Quanto mais o histograma se aproxima de uma curva simétrica em forma de sino, maiores são os indícios de que os dados apresentam distribuição normal. Importante ressaltar que a avaliação visual deve ser complementada por testes estatísticos de normalidade.

**Gráfico Q-Q**

O gráfico Q-Q compara os quantis observados nos dados com os quantis esperados de uma distribuição normal teórica. Quando os pontos se distribuem próximos a uma linha reta, há indícios de que os dados apresentam normalidade. Por outro lado, os desvios podem indicar assimetria, presença de valores extremos ou outras diferenças em relação à distribuição normal.

![Gráfico Q_Q](https://preview.redd.it/interpreting-qq-plot-v0-7gylndiuqg181.png?width=640&crop=smart&auto=webp&s=a314d931ee7a5f73e5c7728e137dd51d5a78bd6c)

Trata-se de um método mais sensível que o histograma, o que o torna especialmente útil para identificar desvios sutis da normalidade.

**Box plot**

O box-plot é um gráfico que resume a distribuição dos dados por meio da mediana, dos quartis e dos valores extremos. Ele permite avaliar a dispersão dos dados, identificar possíveis outliers e verificar indícios de simetria. Em uma distribuição aproximadamente normal, a mediana tende a ficar próxima ao centro da caixa e os bigodes apresentam comprimentos semelhantes em ambos os lados.

![Boxplot](https://midias.siteware.com.br/wp-content/uploads/2024/04/20163554/BoxPlot_1.jpg)

### Testes para normalidade

Os testes analíticos mais utilizados para avaliar a normalidade dos dados incluem o teste de Shapiro-Wilk, que é mais indicado para amostras pequenas (< 50), e o teste de Kolmogorov-Smirnov, mais utilizado em amostras maiores (>50). Além desses, também podem ser aplicados outros testes, como Anderson-Darling, Cramer-von Mises e Lilliefors.

A interpretação desses testes é feita a partir do p-valor. De forma geral, quando o p-valor é maior que 0,05, considera-se que os dados apresentam comportamento compatível com uma distribuição normal, ou seja, não há evidências para rejeitar essa condição. Por outro lado, quando o p-valor é menor ou igual a 0,05, entende-se que os dados não seguem uma distribuição normal.
